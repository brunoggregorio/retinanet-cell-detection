```python
# show images inline
%matplotlib inline

from matplotlib import pyplot as plt
from utils.point_spread_functions import createPSFKernel
```

```python
# Create kernel with Airy disk pattern
kernel = createPSFKernel(k_type=1, k_param=45)

# Show kernel created
plt.imshow(kernel, interpolation='none', origin='lower', cmap='gray')
plt.xlabel('x [pixels]')
plt.ylabel('y [pixels]')
plt.colorbar()
plt.show()
```

```python
# Create Gaussian kernel
kernel = createPSFKernel(k_type=0, k_size=49, k_param=9.0)

# Show kernel created
plt.imshow(kernel, interpolation='none', origin='lower', cmap='gray')
plt.xlabel('x [pixels]')
plt.ylabel('y [pixels]')
plt.colorbar()
plt.show()
```
